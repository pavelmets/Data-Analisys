###### Считать и вывести изображение

```python
import cv2
from matplotlib import pyplot as plt
import numpy as np

#вывод изображения
#функция cv2.imread импортирует изображение в массив numpy array в формате BGR.
image = cv2.imread('pictures/big.jpg') 
plt.imshow(image)
```

```python
#выведем представление изображения
image
```

```python
#посмотрим на размерность изображения image
image.shape
```

###### Какого размера изображение?

```python
#нахождение размера изображения (произведение ширины на высоту)
image_size = image.shape[0]*image.shape[1]
print(f'Размер изображения {image_size} пикселей')
```

###### Найдите среднее значение по каждому из каналов (среднее в 3 канале, округлить до целого)

```python
vector_blue = image[:, :, 0].flatten()
vector_green = image[:, :, 1].flatten()
vector_red = image[:, :, 2].flatten()
mn_blue = vector_blue.mean().round()
mn_green = vector_green.mean().round()
mn_red = vector_red.mean().round()
print(f'Среднее значение по синему каналу - {mn_blue}', f'Среднее значение по зеленому каналу - {mn_green}', f'Среднее значение по красному каналу - {mn_red}', sep='\n')
```

###### Посчитайте разброс(среднеквадратичное отклонение) значений по каждому из каналов (разброс в 1 канале, округлить до целого)

```python
blue_D = vector_blue.std().round()
green_D = vector_green.std().round()
red_D = vector_red.std().round()
print(f'Разброс значений по синему каналу - {blue_D}')
print(f'Разброс значений по зеленому каналу - {green_D}')
print(f'Разброс значений по красному каналу - {red_D}')
```

###### Поменяйте каналы исходного изображения в обратном порядке, какое число стоит на позиции [0,345,0]?

```python
#функция np.flip меняет местами элементы вдоль указанной оси axis
changed_image = np.flip(image, axis=2) 
plt.imshow(changed_image)
print(f'На позиции  [0,345,0] стоит число {changed_image[0, 345, 0]}')
```

###### Создадим новую матрицу размером 500х500. Сколько столбцов, в которых есть хоть одно значение 7?

```python
np.random.seed(7964)
A = np.random.randint(0, 8, size=(500, 500))
count = np.any(A==7, axis=1).sum()
print("Количество столбцов, где есть хоть одна 7:", count)
```

###### Найдите индексы значений, которые меньше 4. Ответ - сумма сумм индексов строк и столбцов между собой.

```python
i,j = np.where(A<4)
result = i.sum() + j.sum()
print('Cумма сумм индексов строк и столбцов между собой: ', result)
```

###### Найдите индексы значений, которые меньше 4 и не меньше 2. Ответ - сумма сумм индексов строк и столбцов между собой.

```python
mask = np.logical_and(A >= 2, A < 4)
i = np.where(mask)[0]
j = np.where(mask)[1]
result = i.sum() + j.sum()
print('Cумма сумм индексов строк и столбцов между собой: ', result)
```

###### Просуммировать все значения по индексам значений из предыдущего задания.

```python
summa = np.sum(A[mask])
print("Сумма значений по индексам предыдущего задания:", summa)
```

###### Найти максимальное значение в каждом столбце массива. Ответ - сумма значений.

```python
sum = np.max(A, axis=0).sum()
print("Сумма всех максимальных значений в каждом столбце массива:", sum)
```

###### Найти скалярное произведение массива A и x. Ответ - сумма значений.

```python
#задание массива х
x = np.arange(1, 501, 1)
scalar1 = np.dot(x, A)
sum1 = np.sum(scalar1)
print('Скалярное произведение А и х:', sum1)
```

###### Умножить массив A на массив w транспонированный. Ответ - сумма значений.

```python
np.random.seed(33)
#задание матрицы w
w = np.random.randint(0, 8, size=(5, 500))
scalar2 = np.matmul(A, w.T)
sum2 = np.sum(scalar2)
print('Скалярное произведение А и w:', sum2)
```
